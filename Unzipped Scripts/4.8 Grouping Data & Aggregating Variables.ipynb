# Aggregating Data with agg()

```python
# Performing single aggregating
df.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
# Performing multiple aggregations 
df.groupby('department_id').agg({'order_number' : ['mean', 'min', 'max']})
```

# Aggregating Data with transform()

```python
# Creating 'max_order' column by finding max 'order_number' for each customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking DataFrame
ords_prods_merge.head(15)
```

```python
# Creating new loyalty flag with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Checking Value Count of 'loyalty_flag'
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking specific Columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# Question 6

```python
# Creating new column to find the mean price using transform()
ords_prods_merge['average_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking DataFrame
ords_prods_merge.head(10)
```

```python
#Creating Spending Flag
ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spending'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# Checking value counts
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Question 6

```python
# Creating new column to find the median days_since_prior_order using transform()
ords_prods_merge['median_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking DataFrame
ords_prods_merge.head()
```

```python
# Creating frequency flag using loc()
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] > 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_order'] > 10) & (ords_prods_merge['median_days_since_order'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Checking frequency_flag value counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
