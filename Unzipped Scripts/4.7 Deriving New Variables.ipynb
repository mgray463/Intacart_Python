```python
# Importing DataFrame
df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# Subsetting the first 1 million entries
df = df_ords_prods[:1000000]
```

# Creating If-Statement

```python
def price_label (row):
    
    if row ['prices'] <= 5:
        return 'Low-Range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-Range product'
    elif row['prices'] > 15:
        return 'High-Range product'
    else: return np.nan
```

## Applying the If-Statement

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Checking the value counts of the Price Range column
df['price_range'].value_counts(dropna = False)
```

# Creating a loc() function

```python
# Creating condition using loc()
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <=5, 'price_range_loc'] = 'Low range product'
```

```python
# Value count for 'price_range_loc'
df['price_range_loc'].value_counts(dropna = False)
```

# Applying loc() function to entire DataFrame

# For-Loops

```python
# creating a value count for orders day of the week
df_ords_prods['order_dow'].value_counts(dropna = False)
```

```python
# creating for-loop
result = []

for value in df_ords_prods['order_dow']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# checking 'result' list
result
```

```python
# Merging list with df_ords_prods dataframe
df_ords_prods['busiest_day'] = result
```

```python
# Checcking the value count for new column
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

# Question 2: Creating 'busiest days" column

```python
# Following same for-loop process as above, except adding another 'busiest day' and 'least busy' day 
result_1 = []

for value in df_ords_prods['order_dow']:
    if value == 0:
        result_1.append('Busiest day')
    elif value == 1:
        result_1.append('Busiest day')
    elif value == 4:
        result_1.append('Least busy')
    elif value == 3:
        result_1.append('Least busy')
    else:
        result_1.append('Regularly busy')
```

```python
# Checking list
result_1
```

```python
# Merging list result_1 with dataframe
df_ords_prods['busiest_days'] = result_1
```

```python
# Drop 'busiest day' column
df_ords_prods = df_ords_prods.drop(columns = ['busiest_day'])
```

```python
# Checking value counts of new column
df_ords_prods['busiest_days'].value_counts(dropna = False)
```
